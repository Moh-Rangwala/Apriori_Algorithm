```python
######## CODE FOR TASK1 ###########
bread_orig = pd.read_csv('bread basket.csv') ### read data from original dataset
bread_orig.columns= ['date_time','Transaction','Item','period_day','weekday_weekend']
filter_data = bread_orig.groupby('Transaction')['Item'].apply(','.join).reset_index() ## filter data and group by transaction ID
filter_data = filter_data.drop(['Transaction'], axis = 1)
print(filter_data)
```

```python
bread_dataset= pd.DataFrame(filter_data['Item'].str.split(',').tolist())
bread_dataset.to_csv('bread_basket_dataset.csv', index = False,header=False) ## convert filtered data to new dataset csv
new_data = pd.read_csv('bread_basket_dataset.csv', header=None) ## read new data
print("First seven transactions are:\n")
new_data.head(7)
```
